# 1 Dataset Configuration

Before we get started, we need to configure the provided data files in order to make them compatible with the format expected by our neural net

Let's rename our training images for clarity.

```python
import os
path = "./data/train_photos/"
filenames = os.listdir(path)

for filename in filenames:
    if filename[0] == '.':
        os.remove(path+filename)
    else:
        os.rename(path+filename, path+"train"+filename)
```

```python
# helper function to replace numeric labels with word labels
mapping = {"0" : "zero", "1" : "one", "2" : "two", "3" : "three", "4" : "four", "5" : "five", "6" : "six", "7" : "seven", "8" : "eight", "9" : "nine"}

def replace(label):
    for c in label:
        if c != " ":
            label = label.replace(c, mapping[c])
    return label
```

```python
import csv

with open('data/train.csv') as biz2label:
    with open('data/train_photo_to_biz_ids.csv') as photo2biz:
        with open('data/train2.csv', 'w') as train:
            reader1 = csv.DictReader(biz2label)
            reader2 = csv.DictReader(photo2biz)

            header_fields = ["image_name", "labels"]
            writer = csv.DictWriter(train, header_fields)
            writer.writeheader()

            biz_to_photo_dict = dict()

            for row2 in reader2:
                biz_to_photo_dict[row2['business_id']] = row2['photo_id']

            for row1 in reader1:
                print("labels", row1['labels'])
                print(replace(row1['labels']))
                writer.writerow({ 'image_name' : 'train_'+str(biz_to_photo_dict[row1['business_id']]),
                                 'labels' : replace(row1['labels']) })
```
